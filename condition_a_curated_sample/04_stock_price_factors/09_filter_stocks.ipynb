```python
import pandas as pd
import numpy as np

# 读取数据，使用GBK编码
df = pd.read_csv(r'C:\Users\31314\Desktop\合并后数据.csv', encoding='gbk')

# 将Mkt-RF列转换为字符串，然后查找任何非数值的行
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

# 检查哪些行不是数字
non_numeric_mask = ~df['Mkt-RF'].apply(lambda x: is_number(str(x)) if pd.notnull(x) else True)

if non_numeric_mask.any():
    print(f"发现 {non_numeric_mask.sum()} 行Mkt-RF列不是数值:")
    print(df.loc[non_numeric_mask, 'Mkt-RF'].unique())
    
    # 删除非数值的行
    df_clean = df[~non_numeric_mask].copy()
    
    # 将Mkt-RF列转换为数值类型
    df_clean['Mkt-RF'] = pd.to_numeric(df_clean['Mkt-RF'], errors='coerce')
    
    print(f"原始数据行数: {len(df)}")
    print(f"清理后行数: {len(df_clean)}")
    print(f"删除行数: {len(df) - len(df_clean)}")
    
    # 保存结果，使用utf-8编码确保兼容性
    df_clean.to_csv(r'C:\Users\31314\Desktop\合并后数据_清理后.csv', index=False, encoding='utf-8')
    print("清理后的数据已保存")
else:
    print("Mkt-RF列所有值都是数值")
    df['Mkt-RF'] = pd.to_numeric(df['Mkt-RF'], errors='coerce')
    df.to_csv(r'C:\Users\31314\Desktop\合并后数据_清理后.csv', index=False, encoding='utf-8')
    print("数据已保存（无清理操作）")
```

```python
# 1. 读取数据
df = pd.read_csv(r"C:\Users\31314\Desktop\合并后数据_清理后.csv")
df['date'] = pd.to_datetime(df['date'])

# 2. 计算每个ticker的数据行数，并过滤掉行数少于372的
filtered_df = df[df.groupby('ticker')['ticker'].transform('count') >= 372]

# 3. 保存结果
output_path = r'C:\Users\31314\Desktop\最终面板数据.csv'
filtered_df.to_csv(output_path, index=False)

print(f"处理完成。")
print(f"原始数据行数: {len(df)}")
print(f"过滤后行数: {len(filtered_df)}")
print(f"结果已保存至: {output_path}")
```
